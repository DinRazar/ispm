### 1.1. Назначение диаграммы вариантов использования

Диаграмма вариантов использования (Use Case Diagram) показывает, **какие типы пользователей (акторы)** взаимодействуют с системой и **какой функционал система им предоставляет**.  
Она отвечает на вопросы:

- кто использует систему;
- какие основные задачи пользователь может выполнить;
- какие сценарии являются базовыми, а какие дополняющими (include/extend).

Для проекта MaRS Land Cover Platform диаграмма вариантов использования опирается на:

- ранее определённые целевые группы пользователей (GIS‑аналитики, агро‑специалисты, администраторы);
- функциональные требования SRS (задание области, загрузка данных, запуск анализа, просмотр результатов, экспорт, администрирование);
- архитектуру и WBS (выделение отдельных подсистем аналитики, загрузки данных и управления пользователями).

Диаграмма помогает связать **концептуальное описание проекта** и **формализованные требования** с архитектурой и реализацией.

### 1.2. Акторы системы

В системе MaRS Land Cover Platform выделяются следующие основные акторы:

- **GIS‑аналитик** – продвинутый пользователь, который:
  - задаёт область интереса и период наблюдения;
  - запускает поиск и загрузку спутниковых снимков;
  - конфигурирует и запускает классификацию земных покровов;
  - анализирует карты и статистику, экспортирует результаты.

- **Агро‑специалист (менеджер по землям)** – прикладной пользователь, который:
  - просматривает готовые карты и статистику по земным покровам;
  - принимает решения на основе полученных результатов;
  - при необходимости экспортирует результаты в удобных форматах.

- **Администратор** – технический пользователь, который:
  - управляет учётными записями и ролями;
  - настраивает параметры системы и доступ к внешним источникам данных;
  - контролирует работоспособность и доступность сервиса.

- **Внешняя система** – интегрируемое приложение (например, корпоративная ГИС или BI‑платформа), которое:
  - обращается к REST API для запуска анализов;
  - получает результаты классификации и статистику в машинно‑читаемом виде.

### 1.3. Основные варианты использования

Минимальный набор ключевых вариантов использования:

1. **UC1: Войти в систему**  
   Пользователь вводит логин и пароль, система выполняет аутентификацию и предоставляет доступ в зависимости от роли (аналитик, агро‑специалист, администратор).

2. **UC2: Задать область и период наблюдения**  
   Пользователь выбирает область интереса (на карте или загрузкой GeoJSON/KML) и задаёт временной интервал, который будет использоваться для поиска данных и анализа.

3. **UC3: Загрузить спутниковые снимки**  
   Система по заданной области и интервалу ищет подходящие сцены в открытых хранилищах (Copernicus, USGS, GEE) и загружает их в локальное хранилище для дальнейшей обработки.

4. **UC4: Запустить классификацию земных покровов**  
   Пользователь выбирает тип анализа (оптика, SAR, мультимодальный) и модель (например, MaRS или RNN/LSTM), после чего система выполняет предобработку и классификацию полученных снимков.

5. **UC5: Просмотр карты и статистики**  
   Система отображает карту классификации, даёт возможность включать/выключать слои (исходные снимки, индексы, маски) и показывает статистику по классам земных покровов для выбранной области.

6. **UC6: Экспорт результатов**  
   Пользователь экспортирует результаты в форматы GeoTIFF, GeoJSON/KML, GIF/MP4 или отчёт (таблицы, метрики качества) для дальнейшего использования в GIS/отчётности.

7. **UC7: Управление пользователями**  
   Администратор создаёт и блокирует учётные записи, назначает роли, настраивает параметры аутентификации и доступа.

### 1.4. Отношения include и extend

Чтобы диаграмма не была перегруженной, достаточно использовать несколько базовых отношений:

- **UC2 включается (include) в UC3 и UC4**  
  Любая загрузка данных и запуск анализа требуют заранее определённой области и периода.

- **UC3 включается (include) в UC4**  
  Перед запуском классификации должны быть найдены и загружены соответствующие спутниковые сцены.

- **UC5 расширяет (extend) UC4**  
  После завершения классификации пользователь может перейти к просмотру карты и статистики.

- **UC6 расширяет (extend) UC5**  
  Экспорт результатов выполняется по инициативе пользователя, когда он уже просматривает карту и статистику.

### 2.2. Основные классы и их назначение

**Класс `User`**

- Представляет пользователя системы.
- Основные атрибуты:
  - `id` – уникальный идентификатор;
  - `username` – логин;
  - `passwordHash` – хэш пароля;
  - `email` – e‑mail;
  - `role` – роль пользователя (например, `ANALYST`, `VIEWER`, `ADMIN`).
- Основные методы:
  - `authenticate(pwd)` – проверка пароля;
  - `canExecute(useCase)` – проверка права на выполнение сценария.

**Класс `AnalysisRequest`**

- Представляет запрос на выполнение анализа (классификации).
- Основные атрибуты:
  - `id` – идентификатор запроса;
  - `aoi` – геометрия области интереса;
  - `dateFrom`, `dateTo` – временной интервал;
  - `analysisType` – тип анализа (оптический, SAR, мультимодальный);
  - `modelName` – название используемой модели (MaRS, RNN, LSTM);
  - `status` – состояние запроса (`PENDING`, `RUNNING`, `DONE`, `FAILED`).
- Основные методы:
  - `validate()` – проверка корректности входных параметров;
  - другие сервисные методы (например, смена статуса).

**Класс `SatelliteScene`**

- Описывает отдельную спутниковую сцену.
- Основные атрибуты:
  - `id` – идентификатор сцены в каталоге;
  - `platform` – платформа (`Sentinel-2`, `Sentinel-1`, `Landsat`);
  - `acquisitionDate` – дата съёмки;
  - `cloudCoverage` – облачность;
  - `downloadUrl` – ссылка на скачивание.
- Основные методы:
  - `download()` – загрузка сцены;
  - `preprocess()` – предобработка (маска облаков, нормализация и т.д.).

**Класс `ClassificationResult`**

- Представляет результат классификации для конкретного запроса.
- Основные атрибуты:
  - `id` – идентификатор результата;
  - `rasterPath` – путь к GeoTIFF с результирующей классификацией;
  - `createdAt` – дата создания результата.
- Основные методы:
  - `getArea(classCode)` – вычисление площади для заданного класса;
  - дополнительные методы экспорта/преобразования.

**Класс `QualityMetrics`**

- Представляет метрики качества классификации.
- Основные атрибуты:
  - `overallAccuracy` – общая точность (OA);
  - `kappa` – коэффициент каппа;
  - опционально – `perClassIoU` (IoU по классам).
- Основные методы:
  - `compute(confusionMatrix)` – расчёт метрик по матрице ошибок.

### 2.3. Связи между классами

- `User` **создаёт** один или несколько `AnalysisRequest`  
  (ассоциация «один ко многим»).

- `AnalysisRequest` **связан** с одним или несколькими `SatelliteScene`

- `AnalysisRequest` **содержит** (композиция) один `ClassificationResult`

- `ClassificationResult` **содержит** (композиция) один `QualityMetrics`  
  – метрики качества считаются для конкретного результата.

### 4.2. Ключевой сценарий: запуск классификации земных покровов

Рассматриваем сценарий: **«Пользователь запускает классификацию через веб‑интерфейс»**.

#### Участники

- `User` – GIS‑аналитик.
- `Web Client` – веб‑интерфейс.
- `API Gateway` – точка входа REST API.
- `Analysis Service` – сервис анализа.
- `DB (PostGIS)` – база данных.
- `Data Source` – внешний источник данных (Copernicus/USGS/GEE).
- `ML Model` – исполнение модели (MaRS / RNN/LSTM).
- `Raster Storage` – хранилище растров.

#### Последовательность шагов

1. Пользователь в `Web Client` задаёт область интереса и период наблюдения, выбирает модель и нажимает кнопку **«Запустить анализ»**.
2. `Web Client` отправляет в `API Gateway` запрос `POST /analysis` с параметрами (`aoi`, `dateFrom`, `dateTo`, `analysisType`, `modelName`).
3. `API Gateway` создаёт новую запись `AnalysisRequest` в `DB` со статусом `PENDING`.
4. `API Gateway` передаёт команду `startAnalysis(requestId)` в `Analysis Service` (как правило, асинхронно).
5. `Analysis Service` обновляет статус запроса в `DB` на `RUNNING`.
6. `Analysis Service` вызывает `Data Source` (интеграция с Copernicus/USGS/GEE) для поиска сцен по заданным параметрам.
7. `Data Source` возвращает список подходящих сцен; `Analysis Service` инициирует их скачивание и сохраняет файлы в `Raster Storage`, а метаданные – в `DB`.
8. `Analysis Service` запускает `ML Model`, передавая ей параметры и пути к данным в `Raster Storage`.
9. `ML Model` считывает входные данные из `Raster Storage`, выполняет предобработку и классификацию, затем записывает результирующий GeoTIFF и, при необходимости, анимацию (GIF/MP4).
10. `ML Model` возвращает в `Analysis Service` информацию о результатах (пути к файлам, базовая статистика).
11. `Analysis Service` создаёт запись `ClassificationResult` и `QualityMetrics` в `DB` и обновляет статус запроса на `DONE`.
12. Параллельно или позже `Web Client` периодически запрашивает статус `GET /analysis/{id}` у `API Gateway`.
13. `API Gateway` читает статус и результаты из `DB` и возвращает их в `Web Client`.
14. `Web Client` отображает карту и статистику пользователю; при необходимости пользователь запускает экспорт.

## 5. Обоснование выбранной архитектуры и структуры компонентов

Выбранная архитектура разделяет систему на три уровня: **представление (Presentation)**, **бизнес‑логика (Business Logic)** и **доступ к данным (Data Access)**. Такой подход позволяет независимо развивать и масштабировать интерфейс, вычислительные сервисы и хранилища данных. Веб‑клиент реализует UX/UI‑часть с учётом ранее проанализированных принципов (законы Хика, Фиттса, Миллера, стандарты ISO 9241‑11 и WCAG), а вся ресурсоёмкая обработка спутниковых данных и ML‑моделей вынесена на серверную сторону.

Бизнес‑логика разбита на отдельные компоненты: `API Gateway`, `Analysis Service` и `Auth Service`. Это соответствует декомпозиции в WBS и SRS: независимые модули для интеграции с внешними источниками (Copernicus, USGS, GEE), для запуска моделей MaRS/RNN/LSTM и для управления пользователями. Такая структуризация упрощает поддержку, тестирование и смену конкретных реализаций (например, замену модели или добавление нового источника данных) без изменения внешнего API и пользовательского интерфейса.

Слой доступа к данным (`PostgreSQL/PostGIS` + `Raster Storage`) отражает требования к хранению пространственных и растровых данных (GeoJSON/KML, GeoTIFF, временные серии, анимации) и обеспечивает совместимость с существующими GIS‑инструментами. Внешние компоненты (Copernicus/USGS/GEE) явно выделены как отдельные акторы и компоненты, что подчёркивает зависимость системы от открытых спутниковых данных и готовых карт (Dynamic World, ESA WorldCover) и связывает текущие диаграммы с ранее разработанным анализом аналогов и спецификацией требований.
