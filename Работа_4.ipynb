## **Введение**

**Спецификация требований к программному обеспечению (Software Requirements Specification, SRS)** является одним из основных документов в управлении проектами создания информационных систем. Этот документ служит связующим звеном между заказчиком и разработчиками, определяя четкое понимание того, что должна делать система, каким требованиям она должна соответствовать и как будет оцениваться успешность проекта.

Согласно статистике Standish Group, около 70% IT-проектов терпят неудачу из-за неправильно определенных или непонятых требований.

Грамотно составленная спецификация требований позволяет:
- Снизить риски проекта за счет раннего выявления противоречий и неясностей
- Обеспечить общее понимание целей между всеми заинтересованными сторонами
- Создать основу для планирования, разработки и тестирования системы
- Предотвратить "расползание границ проекта" (scope creep)
- Сформировать критерии приемки системы

# СПЕЦИФИКАЦИЯ ТРЕБОВАНИЙ К ПРОГРАММНОМУ ОБЕСПЕЧЕНИЮ

## 1. ВВЕДЕНИЕ

### 1.1. Цель документа
Настоящий документ определяет полный набор требований к информационной системе для классификации земного покрытия по мультимодальным спутниковым данным (оптическим и радарным). Документ предназначен для согласования между разработчиком, потенциальными пользователями и руководителем дипломного проекта, а также служит основой для последующего проектирования, разработки и тестирования системы.

### 1.2. Область применения
Система предназначена для автоматической классификации земного покрытия по пяти категориям (вода, снег, растительность, гарь, обнажённый грунт) на основе совместной обработки оптических снимков (Sentinel-2, Landsat) и радарных снимков (Sentinel-1). Система также предоставляет возможность анализа временных рядов и краткосрочного прогнозирования динамики земного покрытия.

Система будет использоваться следующими категориями пользователей:

- Исследователи в области экологии, географии и природопользования

- Студенты профильных направлений

- Специалисты по мониторингу природных территорий

- Государственные природоохранные организации

Система реализована в виде веб-приложения с интерактивной картой и не требует от пользователя навыков программирования.

### 1.3. Определения, акронимы и сокращения
- ДЗЗ — Дистанционное зондирование Земли

- SRS — Software Requirements Specification (Спецификация требований к программному обеспечению)

- MaRS — архитектура нейронной сети для мультимодальной классификации (оптика + радар)

- RNN — Рекуррентная нейронная сеть (Recurrent Neural Network)

- LSTM — Long Short-Term Memory (тип рекуррентной нейронной сети)

- NDVI — Normalized Difference Vegetation Index (вегетационный индекс)

- SAR — Synthetic Aperture Radar (радар с синтезированной апертурой)

- GEE — Google Earth Engine

- GeoJSON — формат хранения географических данных

- KML — Keyhole Markup Language (формат для отображения географических данных)

- GeoTIFF — формат хранения растровых географических данных

- Классификация земного покрытия — процесс отнесения каждого пикселя спутникового снимка к одному из типов поверхности (вода, снег, растительность, гарь, грунт)

- Временной ряд — последовательность значений одного показателя за разные моменты времени

- Прогноз — предсказание будущих значений на основе анализа исторических данных

### 1.4. Пользовательские роли

Основные задачи:

- Анализ временных рядов изменений

- Формирование прогнозов

- Экспорт результатов для дальнейшего анализа

**Студент**

Основные задачи:

## 2. ФУНКЦИОНАЛЬНЫЕ ТРЕБОВАНИЯ

### 2.1. Управление данными и загрузка снимков

**ФТ-001: Выбор территории на карте**

- Описание: Пользователь должен иметь возможность выбрать интересующий участок земной поверхности на интерактивной карте для последующего анализа

- Входные данные: Координаты клика или выделенная прямоугольная область на карте

- Выходные данные: Отображение выбранной области с возможностью уточнения границ

- Бизнес-правила: Минимальный размер выбираемой области — 1 кв. км, максимальный — 100 кв. км (ограничение вычислительных ресурсов)

- Приоритет: Критический

- Критерий проверки: Пользователь может выделить область на карте, система корректно отображает границы выбранного участка

**ФТ-002: Загрузка пользовательских границ участка**

- Описание: Пользователь должен иметь возможность загрузить границы интересующего участка в форматах GeoJSON или KML для анализа конкретного земельного участка

- Входные данные: Файл в формате .geojson или .kml

- Выходные данные: Отображение загруженных границ на карте

- Бизнес-правила: Система проверяет валидность геометрии (замкнутый полигон). При невалидной геометрии выдается сообщение об ошибке

- Приоритет: Высокий

- Критерий проверки: Пользователь загружает файл с границами, система отображает полигон на карте

**ФТ-003: Автоматическая загрузка спутниковых снимков**

- Описание: Система должна автоматически загружать актуальные оптические (Sentinel-2, Landsat) и радарные (Sentinel-1) снимки для выбранной территории из открытых архивов Copernicus Open Access Hub

- Входные данные: Координаты и границы выбранной территории, диапазон дат (по умолчанию — последний месяц)

- Выходные данные: Массив загруженных и предобработанных снимков

- Бизнес-правила: Загрузка выполняется в фоновом режиме с отображением прогресса. При недоступности спутниковых данных выводится сообщение с предложением выбрать другой период

- Приоритет: Критический

- Критерий проверки: Для выбранной территории система успешно загружает снимки за указанный период

**ФТ-004: Предобработка спутниковых снимков**

- Описание: Система должна выполнять автоматическую предобработку загруженных снимков для подготовки к классификации

- Входные данные: Сырые оптические и радарные снимки

- Выходные данные: Предобработанные снимки, готовые для подачи в модель классификации

Функциональность предобработки:

- Обрезание снимков по границам выбранной территории

- Маскирование облаков для оптических снимков

- Калибровка радарных снимков (калибровка, фильтрация спеклов)

- Ресэмплинг к единому разрешению (10 м)

- Нормализация значений

- Приоритет: Критический

- Критерий проверки: Предобработка выполняется успешно, выходные данные имеют корректные размеры и значения

### 2.2. Классификация земного покрытия

**ФТ-005: Классификация земного покрытия по оптическим данным**

- Описание: Система должна выполнять классификацию земного покрытия на основе только оптических спутниковых снимков (Sentinel-2, Landsat) при наличии данных

- Входные данные: Предобработанный оптический снимок

- Выходные данные: Карта классификации с указанием для каждого пикселя одного из пяти классов: вода, снег, растительность, гарь, обнажённый грунт

- Бизнес-правила: Классификация выполняется с использованием обученной сверточной нейронной сети

- Приоритет: Критический

- Критерий проверки: Для тестового участка с известным типом покрытия классификация соответствует реальности с точностью не менее 85%

**ФТ-006: Классификация земного покрытия по радарным данным**

- Описание: Система должна выполнять классификацию земного покрытия на основе только радарных снимков (Sentinel-1) для обеспечения работы в условиях облачности

- Входные данные: Предобработанный радарный снимок (VV и VH поляризации)

- Выходные данные: Карта классификации с указанием для каждого пикселя одного из пяти классов

- Бизнес-правила: Классификация выполняется с использованием обученной сверточной нейронной сети, адаптированной к SAR-данным

- Приоритет: Критический

- Критерий проверки: В условиях облачности (когда оптические снимки недоступны) система выдает классификацию на основе радарных данных

**ФТ-007: Мультимодальная классификация (оптика + радар)**

- Описание: Система должна выполнять совместную классификацию земного покрытия на основе объединения оптических и радарных данных с использованием архитектуры MaRS для повышения точности и устойчивости

- Входные данные: Пара предобработанных снимков (оптический + радарный) за одну дату

- Выходные данные: Итоговая карта классификации с повышенной точностью

- Бизнес-правила: При наличии обоих типов данных приоритет отдается мультимодальной классификации. Если доступен только один тип данных, используется соответствующая модель

- Приоритет: Критический

- Критерий проверки: Точность мультимодальной классификации выше точности классификации по отдельному типу данных не менее чем на 5%

**ФТ-008: Отображение результатов классификации на карте**

- Описание: Система должна отображать результаты классификации в виде цветовой карты на интерактивной карте

- Входные данные: Результат классификации (матрица классов)

- Выходные данные: Наложенный слой на карте, где каждому классу соответствует свой цвет (вода — синий, снег — голубой, растительность — зелёный, гарь — коричневый, грунт — жёлтый)

- Бизнес-правила: Должна быть возможность включать/выключать отображение слоя классификации, а также регулировать его прозрачность

- Приоритет: Критический

- Критерий проверки: Результаты классификации корректно отображаются на карте, цвета соответствуют классам

### 2.3. Анализ временных рядов и прогнозирование

**ФТ-009: Построение временных рядов**

- Описание: Пользователь должен иметь возможность просмотреть, как менялось земное покрытие на выбранной территории за определенный период времени

- Входные данные: Выбранная территория, период времени (от 1 месяца до 2 лет), выбранный класс покрытия (или все классы)

- Выходные данные: График изменения площади каждого типа покрытия во времени, а также анимация изменений на карте

- Бизнес-правила: Временной ряд строится на основе исторических спутниковых данных. Минимальный шаг по времени — 1 месяц

- Приоритет: Высокий

- Критерий проверки: Для выбранной территории и периода система строит график изменения площадей, данные соответствуют реальным спутниковым наблюдениям

**ФТ-010: Краткосрочное прогнозирование**

- Описание: Система должна формировать краткосрочный прогноз (до 30 дней) динамики земного покрытия на основе анализа временных рядов с использованием рекуррентной нейронной сети (RNN/LSTM)

- Входные данные: Выбранная территория, горизонт прогноза (дней, от 7 до 30)

- Выходные данные: Прогнозные карты земного покрытия на заданные даты, график прогнозируемых изменений площадей, показатель уверенности прогноза

- Бизнес-правила: При показатели уверености прогноза ниже 70% выводится предупреждение о пониженной достоверности. Прогноз носит вероятностный характер

- Приоритет: Высокий

- Критерий проверки: Прогноз формируется за время не более 30 секунд, результаты визуализируются на карте

- Входные данные: Временной ряд за выбранный период

- Выходные данные: Видеофайл или GIF-анимация, показывающая динамику изменений

- Приоритет: Средний

- Критерий проверки: Скачанная или отображённая анимация корректно воспроизводится и отображает изменения

## 3. НЕФУНКЦИОНАЛЬНЫЕ ТРЕБОВАНИЯ

### 3.1. Требования к производительности

**НФТ-П-001: Время классификации**

Классификация выбранной территории (площадью до 10 кв. км) должна выполняться не более чем за 30 секунд с момента отправки запроса до отображения результата

**НФТ-П-002: Время построения временного ряд**а

Построение временного ряда за период до 2 лет должно выполняться не более чем за 45 секунд

**НФТ-П-003: Время прогнозирования**
