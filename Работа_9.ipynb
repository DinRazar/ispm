# СТРАТЕГИЯ ТЕСТИРОВАНИЯ И ТЕСТ-КЕЙСЫ

## 1. СТРАТЕГИЯ ТЕСТИРОВАНИЯ

### 1.1. Цели и задачи

Цель тестирования информационной системы MaRS Land Cover Platform — подтвердить, что реализованный функционал выбора области интереса, подбора спутниковых сцен, запуска анализа и просмотра результатов соответствует спецификации требований  и не содержит критических дефектов при работе с реальными геоданными.
Основными задачами тестирования являются: проверка корректности ключевых пользовательских сценариев , проверка стабильности веб‑интерфейса при взаимодействии с картой и протоколами загрузки/экспорта данных, а также верификация основных нефункциональных характеристик (производительность, юзабилити, устойчивость к ошибкам ввода).

### 1.2. Область тестирования

**Включено:**
- Веб‑интерфейс единой рабочей панели анализа земных покровов (главный дашборд, блок «Активный проект», карта анализа, KPI, легенда, блоки «Новый анализ», «Карта и выбор области», «Спутниковые сцены», «Результаты классификации», «Экспорт результатов»).

- Функционал выбора AOI на карте (выделение области, отображение координат, сохранение AOI).

- Подбор и отображение списка спутниковых сцен по заданному интервалу и источнику данных.

- Запуск анализа и получение результирующего слоя классификации, метрик качества и статистики по классам.

- Экспорт результатов в форматы GeoTIFF, GeoJSON/KML и CSV (как предусмотрено требованиями).

**Исключено:**
- Внутренняя ML‑логика моделей классификации (точность алгоритма проверяется на уровне метрик, но не является объектом модульного тестирования в данном задании).

- Низкоуровневые компоненты инфраструктуры (кластер хранения, внешние сервисы поставщиков спутниковых данных), которые рассматриваются как внешние зависимости.

### 1.3. Виды тестирования

- Функциональное тестирование пользовательских сценариев (web UI + серверные API по запуску анализа и экспорту).
- Интеграционное тестирование цепочки «выбор AOI → подбор сцен → запуск анализа → результат → экспорт».
- Регрессионное тестирование ключевых сценариев после внесения изменений в интерфейс или API.
- Тестирование производительности (оценка времени отклика при загрузке карты, списка сцен и результатов анализа для типовых объёмов данных).
- Тестирование безопасности на уровне аутентификации/авторизации (проверка доступа разных ролей к функциям анализа и экспорта).
- Юзабилити‑тестирование интерфейса рабочей панели (понятность блоков, логика движения по экрану сверху вниз, читаемость легенды и KPI).

### 1.4. Инструменты тестирования

| Тип тестирования              | Инструмент                     | Обоснование                                  |
|------------------------------|-------------------------------|----------------------------------------------|
| Функциональное (UI)         | Ручное тестирование в браузере | Интерфейс прототипа реализован как веб‑страница, тестирование сценариев выбора AOI, запуска анализа и экспорта удобно выполнять вручную. |
| Функциональное (API)        | Postman/cURL                  | Позволяет отправлять HTTP‑запросы для запуска анализа и проверки ответов сервера без участия UI. |
| Производительность (basic)  | Встроенные средства браузера (DevTools, вкладка Network) | Дают возможность замерять время загрузки карт и ресурсов при типовых сценариях. |
| Юзабилити                   | Наблюдение по сценариям + опросный лист | Соответствует учебному формату и позволяет получить качественную обратную связь от пользователей. |

### 1.5. Критерии входа и выхода

**Entry Criteria:**
- Базовая функциональность выбора AOI, просмотра сцен, запуска анализа и отображения результатов реализована.
- Прототип интерфейса доступен в стабильной тестовой среде.
- Определён и согласован набор тест‑кейсов и матрица трассируемости для требований с приоритетом «Критический» и «Высокий».
**Exit Criteria:**
- Выполнены все тест‑кейсы для требований с приоритетом «Критический»; не осталось открытых дефектов уровня критический/высокий.
- Покрытие тестами требований с приоритетом «Высокий» составляет не менее 80 %.
- Все дефекты уровня «Средний» либо исправлены, либо задокументированы и запланированы к исправлению в будущих версиях.

### 1.6. Метрики качества

- Покрытие требований тестами (по матрице трассируемости):
  - 100 % для критических требований.
  - ≥ 80 % для требований с приоритетом «Высокий».
- Плотность дефектов: количество найденных дефектов на один тест‑кейс и на один пользовательский сценарий.
- Среднее время ответа интерфейса при загрузке карты анализа и результатов (целевой уровень — не более нескольких секунд для типовых сценариев).
- Доля успешно завершённых тест‑кейсов при первичном прогоне (целевое значение — не менее 90 %).

## 3. НЕФУНКЦИОНАЛЬНЫЕ ТЕСТ-КЕЙСЫ

### 3.1. Тестирование производительности

**TC-PERF-001: Время загрузки рабочей панели**

- Цель: убедиться, что основная панель с ключевыми блоками загружается за приемлемое время при стандартном объёме данных.
- Метрика: время полной загрузки страницы (L), измеряется средствами DevTools.
- Критерий: L не более 5 секунд при первой загрузке страницы.

### 3.2. Тестирование безопасности

**TC-SEC-001: Доступ к функциям анализа только для авторизованных пользователей**

- Цель: проверить, что запуск анализа и экспорт доступны только пользователям с ролью ANALYST/ADMIN.
- Шаги: попытаться открыть блоки «Новый анализ», «Экспорт результатов» без авторизации.
- Ожидаемый результат: интерфейс не предоставляет доступ к запуску анализа и экспорту, система предлагает авторизацию.

### 3.3. Тестирование юзабилити

**TC-UX-001: Понятность структуры единой рабочей панели**

- Цель: оценить, насколько пользователю понятно движение по панели сверху вниз (от активного проекта до экспорта).
- Метод: прохождение типового сценария и анкетирование.
- Критерий: не менее 80 % респондентов оценивают структуру как «понятную» или «очень понятную».
