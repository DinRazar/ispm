#### **2. Ожидаемые результаты проекта**



Опишите конкретные результаты, которые планируется получить по завершении проекта. Какой программный продукт, методика или информационная система будет создан/создана в результате вашей работы?

По завершении проекта будет создана прототип информационной системы (веб-приложение) для оценки пригодности земельных участков на основе анализа спутниковых снимков.

**Программный продукт:**

Веб-приложение с интерактивной картой, позволяющее пользователю выделить участок и получить по нему аналитический отчет.

**Алгоритмы и модели:**

Набор обученных моделей компьютерного зрения для классификации спутниковых снимков (распознавание типов почв, растительности, водных объектов, снежного покрова) и расчета специализированных коэффициентов (например, вегетационный индекс NDVI, индекс влажности NDMI).

**Модуль динамического анализа:**

Инструмент, строящий временные ряды изменений рельефа и состояния участка (графики и анимации изменения снежного покрова, уровня затопления, "выгорания" растительности) за выбранный период.

**Методика расчета:**

Документированная методика интегральной оценки пригодности участка (итоговый коэффициент/балл), учитывающая комбинацию статических (тип почв, уклон) и динамических (частота затоплений, вегетационные циклы) факторов.

#### **3. Методология использования**


Каким образом предполагается практическое применение разрабатываемого решения? Опишите основные сценарии взаимодействия пользователей с разрабатываемым решением.

Практическое применение решения предполагает два основных сценария взаимодействия:

**Сценарий 1 (Оперативная оценка и мониторинг)**

1. Пользователь заходит в веб-приложение и через интерактивную карту выбирает интересующий его земельный участок (или загружает его границы в формате GeoJSON/KML).

2. Система автоматически подгружает актуальные спутниковые снимки (из архивов Sentinel, Landsat) за последний месяц.

3. Пользователь запускает анализ. Система отображает на карте рассчитанные индексы (NDVI), выделяет зоны переувлажнения и эрозии.

4. Пользователь нажимает кнопку "Динамика" и просматривает, как менялось состояние участка в течение последнего года (например, когда сошел снег и начался паводок).

Результат:
Пользователь принимает решение о том, стоит ли начинать посевные работы на этом поле или необходимо отложить их из-за переувлажнения.

**Сценарий 2 (Сравнительный анализ и отчётность)**

1. Пользователь выбирает несколько участков для сравнения.

2. Система генерирует сводную таблицу с рассчитанными коэффициентами пригодности, включая исторические данные по каждому участку (частота затоплений за 5 лет, средние вегетационные показатели).

3. Пользователь экспортирует полученный отчет в PDF или Excel.

Результат:
чиновник получает объективные данные для включения участка в программу мелиорации; оценщик — аргументированную базу для кадастровой стоимости.

#### **4. Входные данные и их источники**


Какие типы данных будут использоваться в качестве входных данных для вашего решения? В каком формате они будут представлены и из каких источников планируется их получение?

В качестве входных данных будут использоваться открытые данные дистанционного зондирования Земли (ДЗЗ) и векторные данные.

Типы и форматы данных:

**Растровые данные: Мультиспектральные снимки со спутников (в формате GeoTIFF).**

Основной источник: Европейское космическое агентство (спутники Sentinel-2), NASA (спутники Landsat-8/9). Снимки содержат спектральные каналы (видимый, инфракрасный), необходимые для расчетов индексов (NDVI, NDMI и др.).

**Векторные данные: Границы земельных участков и зон.**

Источник: Открытые кадастровые карты (Росреестр), сервисы OpenStreetMap, а также пользовательские загрузки в форматах GeoJSON, Shapefile или KML.

**Продукты обработки ДЗЗ:** Готовые цифровые модели рельефа (ЦМР), например, SRTM (Shuttle Radar Topography Mission), для анализа уклонов и высот.
