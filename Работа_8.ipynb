## 1. ИНФОРМАЦИОННАЯ АРХИТЕКТУРА

### 1.1. Карта навигации (структура системы)

```
Единая рабочая панель
├── Поисковая строка
├── Профиль пользователя
├── Активный проект
│   ├── AOI
│   ├── Период
│   ├── Модель
│   └── Источник данных
├── Карта анализа
├── Ключевые показатели
│   ├── Точность
│   ├── IoU
│   ├── Сцены
│   └── Классы
├── Слои
├── Легенда классов
├── Последние действия
├── Новый анализ
├── Карта и выбор области
│   ├── Параметры AOI
│   ├── Координаты
│   └── Следующий шаг
├── Спутниковые сцены
├── Результаты классификации
│   ├── Результирующая карта
│   ├── Метрики качества
│   ├── Статистика по классам
│   └── Действия
├── Экспорт результатов
├── Пользователи
└── Настройки системы
```

Структура интерфейса построена не как набор отдельных страниц, а как единая рабочая панель, в которой пользователь последовательно проходит все этапы анализа земных покровов: от просмотра активного проекта и выбора области интереса до получения результата классификации и экспорта файлов. Такой подход напрямую соответствует присланному макету, где все ключевые блоки системы расположены на одном длинном экране и образуют логически связанную цепочку действий.

### 1.2. Пользовательские пути

**Сценарий 1: Просмотр текущего состояния проекта**

```
Шаг 1: Открытие рабочей панели → Шаг 2: Просмотр блока «Активный проект» → Шаг 3: Ознакомление с картой анализа и KPI
```

**Сценарий 2: Настройка нового анализа**

```
Шаг 1: Переход к блоку «Новый анализ» → Шаг 2: Ознакомление с требованиями сценария → Шаг 3: Переход к блоку «Карта и выбор области» → Шаг 4: Выделение AOI → Шаг 5: Сохранение области
```

**Сценарий 3: Подбор спутниковых сцен**

```
Шаг 1: Выбор области интереса → Шаг 2: Переход к блоку «Спутниковые сцены» → Шаг 3: Просмотр подобранных сцен Sentinel-1 / Sentinel-2 / Landsat → Шаг 4: Выбор оптимальных данных
```

**Сценарий 4: Просмотр результата классификации**

```
Шаг 1: Переход к блоку «Результаты классификации» → Шаг 2: Анализ результирующей карты → Шаг 3: Просмотр метрик качества → Шаг 4: Анализ статистики по классам
```

**Сценарий 5: Экспорт и администрирование**

```
Шаг 1: Переход к блоку «Экспорт результатов» → Шаг 2: Выбор формата GeoJSON / KML / GeoTIFF / CSV → Шаг 3: При необходимости переход к блоку «Пользователи» или «Настройки системы»
```

Основная логика пользовательских путей в этом интерфейсе линейная: пользователь движется сверху вниз по экрану, а каждый следующий блок опирается на предыдущий. Такое решение уменьшает число переключений между страницами и делает рабочий процесс визуально непрерывным.

### 6.2. Применение законов проектирования

**Закон Фиттса:**
ключевые интерактивные блоки достаточно крупные и отделены друг от друга, поэтому пользователю легко попадать по ним и быстро выбирать нужное действие.

**Закон Хика:**
вместо большого количества мелких пунктов пользователю предлагаются крупные смысловые блоки: анализ, выбор области, сцены, результат, экспорт. Это упрощает принятие решений.

**Закон Миллера:**
информация сгруппирована в законченные карточки, и пользователь работает не с десятками разрозненных элементов, а с ограниченным числом логических модулей.

### 6.3. Улучшения относительно аналогов

**Проблема в аналогах:** часто карта, статистика и экспорт разнесены по разным вкладкам.

**Моё решение:** в данном макете они объединены в одну рабочую панель, что уменьшает число переходов.

**Проблема в аналогах:** пользователю сложно понять, на каком проекте он сейчас работает.

**Моё решение:** сверху размещён блок «Активный проект» с AOI, периодом, моделью и источником данных.

**Проблема в аналогах:**  аналитические показатели спрятаны глубоко в интерфейсе.

**Моё решение:** KPI и метрики качества вынесены в отдельные заметные карточки.

**Моё решение:** экспорт оформлен как самостоятельный блок с явным выбором формата.
